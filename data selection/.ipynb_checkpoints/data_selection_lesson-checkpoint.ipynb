## Extract some data

* `FatJet` for jets that are merges
* `Jet` for non-merged jets
* `Muon` for muons
* `PuppiMET` which is missing energy in the transverse plane (MET) for pileup per particle identification (Puppi)



```python
# Fat jets ---------------------------------------------------

# Soft-drop mass, calculated using a particular algorithm
fatjet_mSD = events['FatJet_msoftdrop'].array()

# A newer tagging algorithm to identify merged top-quark jets
fatjet_tag = events['FatJet_particleNet_TvsQCD'].array()

# Measures of subjettiness, used in the original analysis
fatjet_tau2 = events['FatJet_tau2'].array()
fatjet_tau3 = events['FatJet_tau3'].array()

# 4-momentum in pt, eta, phi, mass 
fatjet_pt = events['FatJet_pt'].array()
fatjet_eta = events['FatJet_eta'].array()
fatjet_phi = events['FatJet_phi'].array()
fatjet_mass = events['FatJet_mass'].array()


# Jets ---------------------------------------------------
# B-tagging variable
jet_btag = events['Jet_btagDeepB'].array()

# Measure of quality of measurement of jet
jet_jetid = events['Jet_jetId'].array()

# 4-momentum in pt, eta, phi, mass 
jet_pt = events['Jet_pt'].array()
jet_eta = events['Jet_eta'].array()
jet_phi = events['Jet_phi'].array()
jet_mass = events['Jet_mass'].array()


# Muons ---------------------------------------------------
# Muon isolation
muon_iso = events['Muon_miniIsoId'].array()

# Measure of quality of how well the muon is reconstructed
muon_tightId = events['Muon_tightId'].array()

# 4-momentum in pt, eta, phi, mass 
muon_pt = events['Muon_pt'].array()
muon_eta = events['Muon_eta'].array()
muon_phi = events['Muon_phi'].array()
muon_mass = events['Muon_mass'].array()


# MET ------------------------------------------------------
# 3-momentum in pt, eta, phi, mass 
met_pt = events['PuppiMET_pt'].array()
met_eta = 0*events['PuppiMET_pt'].array()  # Fix this to be 0
met_phi = events['PuppiMET_phi'].array() 

# Scalar quantity used in selection criteria
ht_lep = muon_pt + met_pt
```

## Selections

A series of masks will be created to return only subsets of the data that meet certain criteria. 

Distinction will be made between cuts that select *events* and cuts that select *particles*. For example, an *event* cut might select only events with MET greater than a certain value, since there is only one MET for any event, or events recorded on a Friday. A cut that selects certain particles might choose muons with $p_T$ greater than a certain value or jets tagged as originating from $b$-quark hadronization.

When applying *particle* cuts, only those that also pass the *event* cuts will be selected.

```python
# Particle-specific cuts -------------------------------------------

# Fat jet cuts --------------------------------
# Calculate a ratio of subjettiness variables tau3 and tau2
tau32 = fatjet_tau3 / fatjet_tau2

# Simplified selection criteria for fat jets: 
# Select fat jets with transverse momentum (pt) greater than 500 and 
# particleNet tag value greater than 0.5
cut_fatjet = (fatjet_pt > 500) & (fatjet_tag > 0.5)

# Muon cuts ----------------------------------
# Define selection criteria for muons:
# 1. Transverse momentum (pt) greater than 55
# 2. Pseudorapidity (eta) between -2.4 and 2.4
# 3. Tight identification (tightId) is True
# 4. Isolation value (iso) greater than 1
# 5. Scalar sum of transverse momentum (ht_lep) greater than 150
cut_muon = (muon_pt > 55) & (muon_eta > -2.4) & (muon_eta < 2.4) & \
           (muon_tightId == True) & (muon_iso > 1) & (ht_lep > 150)

# Non-boosted jet cuts ------------------------
# Define selection criteria for non-boosted jets:
# 1. B-tagging score (btagDeepB) greater than 0.5
# 2. Jet identification score (jetId) greater than or equal to 4
cut_jet = (jet_btag > 0.5) & (jet_jetid >= 4)

# Event cuts --------------------------------------------------------
# MET cuts --------------------
# Define selection criteria for missing transverse energy (MET):
# Select events with MET transverse momentum (pt) greater than 50
cut_met = (met_pt > 50)

# Cuts on the number of muons that pass our cuts
# Ensure there is exactly one muon that meets the muon cuts
cut_nmuons = ak.num(cut_muon[cut_muon]) == 1

# Cut on the event passing the trigger
# Select events that pass the 'HLT_TkMu50' trigger
cut_trigger = events['HLT_TkMu50'].array()

# Cut on the number of fat jets that pass our selection criteria
# Ensure there is exactly one fat jet that meets the fat jet cuts
cut_ntop = ak.num(cut_fatjet[cut_fatjet]) == 1

# Combine all the individual cuts (trigger, number of muons, MET, number of fat jets)
cut_full_event = cut_trigger & cut_nmuons & cut_met & cut_ntop
```

## Use the cuts and calculate some values

Create 4-vector objects for the fat jets, jets, muons, and MET. These objects can be used with the Awkward array class to naturally handle combinations of particles within a given event, such as calculating all candidates when more than one jet passes the selection criteria. For MET, assume that the momentum vector represents a neutrino, effectively giving it 0 mass.

```python
# The order of operations is as follows:
# 1. Select events that pass the full event cuts
# 2. Extract the relevant particle properties from the selected events
# 3. Apply specific particle cuts to the selected events to get the final subsets

# Create 4-vector objects for fat jets
fatjets = ak.zip(
    {"pt": fatjet_pt[cut_full_event][cut_fatjet[cut_full_event]],      
     "eta": fatjet_eta[cut_full_event][cut_fatjet[cut_full_event]],     
     "phi": fatjet_phi[cut_full_event][cut_fatjet[cut_full_event]],     
     "mass": fatjet_mass[cut_full_event][cut_fatjet[cut_full_event]]},  
    with_name="Momentum4D",                                             
)

# Create 4-vector objects for muons
muons = ak.zip(
    {"pt": muon_pt[cut_full_event][cut_muon[cut_full_event]],           
     "eta": muon_eta[cut_full_event][cut_muon[cut_full_event]],         
     "phi": muon_phi[cut_full_event][cut_muon[cut_full_event]],        
     "mass": muon_mass[cut_full_event][cut_muon[cut_full_event]]},     
    with_name="Momentum4D",                                             
)

# Create 4-vector objects for jets
jets = ak.zip(
    {"pt": jet_pt[cut_full_event][cut_jet[cut_full_event]],             
     "eta": jet_eta[cut_full_event][cut_jet[cut_full_event]],           
     "phi": jet_phi[cut_full_event][cut_jet[cut_full_event]],           
     "mass": jet_mass[cut_full_event][cut_jet[cut_full_event]]},        
    with_name="Momentum4D",                                             
)

# Create 4-vector objects for MET (missing transverse energy)
met = ak.zip(
    {"pt": met_pt[cut_full_event],                                       
     "eta": met_eta[cut_full_event],                                    
     "phi": met_phi[cut_full_event],                                     
     "mass": 0},                                                         
    with_name="Momentum4D",                                              
)
```

The features of 
[Awkward to calculate the different combinations of particles](https://awkward-array.org/doc/main/getting-started/thinking-in-arrays.html).

This enables the calculation of the sum of the 4-momentum of all particles.

```python
# Calculate all the different combinations
p4mu,p4fj,p4j,p4met = ak.unzip(ak.cartesian([muons, fatjets, jets, met]))

# Calculate a sum of the 4-momenta
p4tot = p4mu + p4fj + p4j + p4met
```

### Calculate the mass and make a quick and simple histogram, just to see if everything work. 

```python
# Get the mass
x = p4tot.mass

print(x)

# Plot 
plt.hist(ak.flatten(x), bins=40, range=(0,4000))
plt.xlabel(f'$M_{{tt}}$ GeV/c$^2$', fontsize=18)
plt.title('Distribution of $M_{tt}$', fontsize=20) 
plt.tight_layout()
plt.savefig('ttbar_mass.png');
```

## Putting it all together

The function processes the input with the following parameters:

- `filename`: the name of the input NanoAOD ROOT file
- `dataset`: a string used to organize output files
- `IS_DATA`: a flag set to `True` if the input data file is collision data

The function writes out `.csv` files using the [`pandas` python library](https://pandas.pydata.org/), saving the following kinematic variables:

- Mass of the $t\overline{t}$ system
- $p_T$ of the muon
- $\eta$ of the muon

These variables will be used for calculating uncertainties later in the analysis. Additionally, the following variables are saved:

- `pileup`: records the number of proton-proton interactions assumed for simulated data
- `weight`: the weight calculated by the Monte Carlo code, often 1 but can vary for each event
- `nevents`: the number of events in the file
- `N_gen`: the difference between `gw_pos` and `gw_neg`, where `gw_pos` is the sum of weights for positive values and `gw_neg` is the sum for negative values. This value is stored for each event entry.

These variables will help in scaling simulation data for comparison with collision data.

When writing out the `.csv` file, include the name of the input NanoAOD file and the dataset name (e.g., `ttsemilep`) in the file name.

```python
def process_file(filename, dataset='default', IS_DATA=False):

    # Print the filename being opened
    print(f"Opening...{filename}")
    
    try:
        # Open the ROOT file
        f = uproot.open(filename)
    except:
        # Print error if file cannot be opened and return None
        print(f"Could not open {filename}")
        return None

    # Access the 'Events' tree from the file
    events = f['Events']

    # Get the number of events in the file
    nevents = events.num_entries

    print(f"{nevents = }")

    # Extract FatJet data
    # FatJet -----------------------------------------------------
    fatjet_mSD = events['FatJet_msoftdrop'].array()

    fatjet_tag = events['FatJet_particleNet_TvsQCD'].array()

    fatjet_tau2 = events['FatJet_tau2'].array()
    fatjet_tau3 = events['FatJet_tau3'].array()

    fatjet_pt = events['FatJet_pt'].array()
    fatjet_eta = events['FatJet_eta'].array()
    fatjet_phi = events['FatJet_phi'].array()
    fatjet_mass = events['FatJet_mass'].array()

    # Extract Muon data
    # Muons -------------------------------------------------------
    muon_pt = events['Muon_pt'].array()
    muon_eta = events['Muon_eta'].array()
    muon_phi = events['Muon_phi'].array()
    muon_mass = events['Muon_mass'].array()

    muon_iso = events['Muon_miniIsoId'].array()

    muon_tightId = events['Muon_tightId'].array()

    # Extract Jet data    
    # Jets -------------------------------------------------------
    jet_btag = events['Jet_btagDeepB'].array()
    jet_jetid = events['Jet_jetId'].array()

    jet_pt = events['Jet_pt'].array()
    jet_eta = events['Jet_eta'].array()
    jet_phi = events['Jet_phi'].array()
    jet_mass = events['Jet_mass'].array()
    
    # Extract MET data
    # MET ---------------------------------------------------------
    met_pt = events['PuppiMET_pt'].array()
    met_eta = 0*events['PuppiMET_pt'].array()  # MET eta is fixed to 0
    met_phi = events['PuppiMET_phi'].array() 

    # Calculate ht_lep as the sum of muon_pt and met_pt
    ht_lep = muon_pt + met_pt
    
    #####################################################################################
    # Cuts
    #####################################################################################

    # Particle-specific cuts --------------------------------------
    tau32 = fatjet_tau3/fatjet_tau2

    cut_fatjet = (fatjet_pt > 500) & (fatjet_tag > 0.5)

    cut_muon = (muon_pt>55) & (muon_eta>-2.4) & (muon_eta<2.4) & \
               (muon_tightId == True) & (muon_iso>1) & (ht_lep>150)

    cut_jet = (jet_btag > 0.5) & (jet_jetid>=4)

    
    # Event cuts -------------------------------------------------
    cut_met = (met_pt > 50)
    cut_nmuons = ak.num(cut_muon[cut_muon]) == 1
    cut_njets = ak.num(cut_jet[cut_jet]) == 1
    cut_trigger = (events['HLT_TkMu50'].array())
    cut_ntop = ak.num(cut_fatjet[cut_fatjet]) == 1

    # Combine all event cuts
    cut_full_event = None
    if IS_DATA:    
        # Build luminosity mask for data
        mask_lumi = build_lumi_mask('Cert_271036-284044_13TeV_Legacy2016_Collisions16_JSON.txt', events)#, verbose=True)
        cut_full_event = cut_trigger & cut_nmuons & cut_met & cut_ntop & mask_lumi
    else:
        cut_full_event = cut_trigger & cut_nmuons & cut_met & cut_ntop
    
    # Apply the cuts and calculate the di-top mass
    fatjets = ak.zip(
        {"pt": fatjet_pt[cut_full_event][cut_fatjet[cut_full_event]], 
         "eta": fatjet_eta[cut_full_event][cut_fatjet[cut_full_event]], 
         "phi": fatjet_phi[cut_full_event][cut_fatjet[cut_full_event]], 
         "mass": fatjet_mass[cut_full_event][cut_fatjet[cut_full_event]]},
        with_name="Momentum4D",
    )

    muons = ak.zip(
        {"pt": muon_pt[cut_full_event][cut_muon[cut_full_event]], 
         "eta": muon_eta[cut_full_event][cut_muon[cut_full_event]], 
         "phi": muon_phi[cut_full_event][cut_muon[cut_full_event]], 
         "mass": muon_mass[cut_full_event][cut_muon[cut_full_event]]},
        with_name="Momentum4D",
    )

    jets = ak.zip(
        {"pt": jet_pt[cut_full_event][cut_jet[cut_full_event]], 
         "eta": jet_eta[cut_full_event][cut_jet[cut_full_event]], 
         "phi": jet_phi[cut_full_event][cut_jet[cut_full_event]], 
         "mass": jet_mass[cut_full_event][cut_jet[cut_full_event]]},
        with_name="Momentum4D",
    )

    met = ak.zip(
        {"pt": met_pt[cut_full_event], 
         "eta": met_eta[cut_full_event], 
         "phi": met_phi[cut_full_event], 
         "mass": 0}, # Assumed mass of neutrino is 0
        with_name="Momentum4D",
    )

    # Combine particle 4-vectors and calculate total momentum
    p4mu,p4fj,p4j,p4met = ak.unzip(ak.cartesian([muons, fatjets, jets, met]))
    p4tot = p4mu + p4fj + p4j + p4met
    
    # Shape the weights and pileup
    # Initialize weights and pileup values
    N_gen = -999
    pileup = -999
    gw_pos = -999
    gw_neg = -999

    pileup_per_candidate = None
    
    tmpval_events = np.ones(len(ak.flatten(p4tot.mass)))
    tmpval = ak.ones_like(p4tot.mass)


    # Put in the MC weights
    # Handle Monte Carlo weights for non-data files
    if not IS_DATA:
        gen_weights = events['genWeight'].array()[cut_full_event]
        pileup = events['Pileup_nTrueInt'].array()[cut_full_event]

        gen_weights_per_candidate = tmpval * gen_weights
        pileup_per_candidate = tmpval * pileup

        # Get values associated with the total number of events. 
        # It's going to duplicate the number of entries, but we'll save the same value to 
        # each event
        gen_weights_org = events['genWeight'].array()

        gw_pos = ak.count(gen_weights_org[gen_weights_org > 0])
        gw_neg = ak.count(gen_weights_org[gen_weights_org < 0])
        N_gen = gw_pos - gw_neg
    else:
        pileup_per_candidate = -999*tmpval
        gen_weights_per_candidate = -999*tmpval
    

    # Build a dictionary and dataframe to write out the subset of data
    # Prepare data for output
    mydict = {}
    mydict['mtt'] = ak.flatten(p4tot.mass) 
    mydict['mu_pt'] = ak.flatten(p4mu.pt) 
    mydict['mu_abseta'] = np.abs(ak.flatten(p4mu.eta))
    mydict['pileup'] = ak.flatten(pileup_per_candidate)
    mydict['weight'] = ak.flatten(gen_weights_per_candidate)
    mydict['nevents'] = nevents*tmpval_events
    mydict['N_gen'] = N_gen*tmpval_events
    mydict['gw_pos'] = gw_pos*tmpval_events
    mydict['gw_neg'] = gw_neg*tmpval_events

    df = pd.DataFrame.from_dict(mydict)

    # Construct output filename and save DataFrame as a CSV file
    outfilename = f"OUTPUT_{dataset}_{filename.split('/')[-1].split('.')[0]}.csv"
    print(f'Saving output to {outfilename}')

    df.to_csv(outfilename, index=False)

    return df
```

Use that function to process a few simulation files and a collision data file.

```python
# Set the dataset name to 'ttleptonic'
dataset = 'ttleptonic'

# Retrieve a list of filenames for the specified dataset. Select one file randomly.
filenames = get_files_for_dataset(dataset, random=True, n=1)

# Get the first filename from the list
filename = filenames[0]

# Process the file to extract and analyze data. Indicate that this is a simulation file (IS_DATA=False).
df = process_file(filename, dataset=dataset, IS_DATA=False)

# Display the resulting DataFrame with the processed data
df
```

```python
# Set the dataset name to 'collision'
dataset = 'collision'

# Retrieve a list of filenames for the 'collision' dataset. Select one file randomly.
filenames = get_files_for_dataset(dataset, random=True, n=1)

# Get the first filename from the list
filename = filenames[0]

# Process the file to extract and analyze data. Indicate that this is a collision data file (IS_DATA=True).
df = process_file(filename, dataset=dataset, IS_DATA=True)

# Display the resulting DataFrame with the processed data
df
```
